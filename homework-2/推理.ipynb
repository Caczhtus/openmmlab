```python
from mmpretrain import ImageClassificationInferencer
import cv2
import matplotlib.pyplot as plt

%matplotlib inline

infer = ImageClassificationInferencer(
    'config/resnet50_1xb256_coslr-100e_fruit-30.py',
    pretrained='ckpt/best_accuracy_top1_epoch_94.pth'
)

lst = ['data/.jpg', 'data/.jpg']

fig, _ = plt.subplots(nrows=2, ncols=1,figsize=(10, 10))
for i, ax in enumerate(fig.axes):
    img = cv2.imread(lst[i])[::-1]
#     ax.axis("off")
    ax.imshow(img)

# plt.savefig("all_test_image.png", bbox_inches='tight')
plt.show()
```

```python
results = infer(lst, batch_size=2, show_dir='./result/')

print_keys = list(results[0].keys())

for i in range(len(lst)):
    print(f"file name: {lst[i]}")
    print(results[i][print_keys[1]])
    print(results[i][print_keys[3]])
    print()

plt.show()
```
